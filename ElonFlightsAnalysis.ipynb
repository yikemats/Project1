```python
import json
```

```python
f = open('sortedElonFlights.json')
flights = json.load(f)
```

```python
flightlist = []

for flight in flights:
    dict = {}
# Duration in hours for the each flight.
    duration_hours = (flight['lastSeen'] - flight['firstSeen'])/3600
# Gulfstream's G650 (Elon's jet) burns right around 452.70 gallons of fuel per hour.
    gallons_per_flight = duration_hours * 452.70
# According to EIA, 9.57 kg of CO2 per gallon.
    kgcarbon_per_flight = gallons_per_flight * 9.57

    dict["date"] = flight['firstSeen']
    dict["departure"] = flight['estDepartureAirport']
    dict["arrival"] = flight['estArrivalAirport']
    dict["carbon"] = round(kgcarbon_per_flight, 2)
    
    flightlist.append(dict)
```

```python
flightlist
```

```python
with open("alldata.json", "w") as output:
    json.dump(flightlist, output)
```

```python
import pandas as pd

df = pd.DataFrame.from_records(flightlist)
```

```python
df.carbon.sum()
```

```python
carbon_totals = []

tracking = 0

for flight in flightlist:
    dict = {}
    dict["date"] = flight["date"]
    carbon = int(flight["carbon"])
    tracking = tracking + carbon
    dict["carbon_cumulative"] = tracking
    carbon_totals.append(dict)
    
print(carbon_totals)
```

```python
df_2 = pd.DataFrame.from_records(carbon_totals)
df_2
```

```python
df.merge(df_2, how="outer", on="date")
df_merged = df.merge(df_2, how="outer", on="date")
```

```python
df.to_csv("final.csv", index=False)
```
